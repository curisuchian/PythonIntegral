```python
while True:                  # Bucle infinito
    try:
        edad = int(input("Por favor ingrese su edad "))
        print("Tiene", edad, "años")
        break
    except ValueError:
        print("Debe ingresar un numero entero!")
```

```python
# Múltiples bloques except
def division(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Division por cero. Por favor corrija.")
    except TypeError:
        print("Tipo de dato equivocado.")
x = 0      
division(1,x)
```

```python
def division(a, b):
    try:
        return a / b
    except:
        print("Hubo error.")
x = 0
division(1,x)
```

```python
while True:
    try:
        x = int(input("Ingrese su edad: "))
    except ValueError:
        print("Por favor ingrese numero entero. Pruebe otra vez...")
    else:
        print('Ahora si,',x,'es un entero')
        break
```

```python
# Interacción con usuario. Uso de finally
veces = 0
while True:
    try:
        x = int(input("Ingrese un numero entero: "))
    except ValueError:
        veces += 1
        print("Numero invalido. Pruebe otra vez...")
    else:
        print('Ahora si,',x,'es un entero')
    finally:
        print("Te equivocaste ", veces, "veces.")
```
