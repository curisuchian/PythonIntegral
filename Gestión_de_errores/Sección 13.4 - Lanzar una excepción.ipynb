#### Lanzar una excepción ('raise an exception')


#### Uso de raise

```python
# La descripción del error aparece con el mensaje de error
x = 5
if x < 10:
    raise ValueError('x debe ser superior a 10.')
```

```python
# Acepta solo numeros reales
def division(a,b):
    if not isinstance(a,float) or not isinstance(b, float):
        raise TypeError('a y b deben ser tipos reales')
    return a / b

division(5.5,6)
```

#### En un bloque try

```python
try:
    a = int(input('Ingrese un número entre 5 y 10: '))
    if a > 10 or a < 5:
        raise ValueError
except ValueError:
    print('Cae fuera del intervalo')
```

#### Uso de assert

```python
def promedio(lista):
    return sum(lista)/len(lista)

assert(promedio([5, 10, 15]) == 10)
print('Todo bien')
```

```python
# Una operacion no admite que la lista ingresada carezca de valores
def promedio(nota):
    assert len(nota) != 0, " La lista esta vacia"
    return sum(nota)/len(nota)

calif = [[4, 6, 7, 4],[40,60,70],[]]
for i in calif:
    print("promedio ", promedio(i))
```

```python
# Asegurar que cierta condición se cumpla.
# Un programa que lee grados Kelvin debe interrumpirse si una temperatura es inferior a 0 grados.
def temp(grados):
    assert (grados > 0), "Inferior al cero absoluto. Detener."
    return grados / 5

temp(-5)
```

#### En una cláusula try except

```python
# Incluir en una cláusula try except, con mensaje
def temp(grados):
    assert (grados > 0), "Inferior al cero absoluto. Detener."
    return grados / 5

try:
    print(temp(-9))
except AssertionError as mensaje:
    print(mensaje)
```
