### Decoradores

```python
def mult(x, y):
    return x * y

mult(4,6)
```

```python
def suma(x, y):
    return x + y

suma(4,6)
```

```python
def aviso(func):
    def advertencia(a,b):
        print('Se advierte que hubo un ingreso a codigo ')
        return func(a,b)
    return advertencia
```

```python
@aviso
def val_abs(x):
    return abs(x)

suma(4,6)
```
