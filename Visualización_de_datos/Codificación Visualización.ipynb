```python
# Ejercicio 1
#!pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 260)
plt.plot(np.sin(x), color='green', linestyle='dotted', linewidth=2.5)
plt.show()
```

```python
# Ejercicio 2
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(100)
y = .4 * x + np.random.randn(100) + 5
#fig, axs = plt.subplots(1,2)   # En fila
fig, axs = plt.subplots(2)   # En columna
axs[0].plot(x, y)
axs[1].plot(x, -y)
plt.show()
```

```python
# Ejercicio 3
import numpy as np
import matplotlib.pyplot as plt

s = np.arange(0., 10., 0.1)
plt.plot(s, s, 'r--', s, s**2, 'bs', s, np.sin(s*5), 'g^')
plt.show()
```

```python
# Ejercicio 4
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.2)
plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo')
plt.subplot(212)
plt.plot(t2, f(t2), 'k')
plt.show()
```
