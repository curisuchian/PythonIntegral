```python
import pandas as pd

datos= [18, 8.5, 8.0, 6.5],[22, 7, 6.5, 4], [20, 3.5, 5, 9]
df=pd.DataFrame(datos)
print(df)
```

```python
# Cambiar indices y encabezamientos. 
df.index = ['alumno 1', 'alumno 2', 'alumno 3']
df.columns = ['edad', 'matematicas', 'economia', 'programacion']
print(df)
```

```python
# Renombrar indices de filas. Indicar eje.
df = df.rename({"alumno 1":"ALUMNO 1","alumno 2":"ALUMNO 2","alumno 3":"ALUMNO 3"}, axis='index')
df
```

```python
# Renombrar encabezamientos. Indicar eje.
df = df.rename({"edad":"Edad","matematicas":"Matematicas","economia":"Economia","programacion":"Programacion"}, 
               axis='columns')
df
```
