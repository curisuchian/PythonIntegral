```python
import pandas as pd

datos_ancho= {'Nombre':['María','Luis','Carmen'],
        'Matemáticas':[8.5, 7, 3.5], 'Economía':[8, 6.5, 5],'Programación':[6.5, 4, 9]}
df=pd.DataFrame(datos_ancho)
df_largo = df.melt(id_vars='Nombre', var_name='Asignatura',value_name='Nota')
print(df_largo)
```

```python
# Volvemos al formato ancho con pivot()
df_ancho = df_largo.pivot(index='Nombre',columns='Asignatura',values='Nota')
print(df_ancho)
```

```python
import pandas as pd

Cordoba = pd.read_csv('Cordoba_precip.csv')
Cordoba
```
