### El m√©todo format()

#### Ordenamiento de los argumentos

```python
print("Numero entero {}".format(42)) 
```

```python
print("Provincia de la Argentina: {}".format('Cordoba'))
```

```python
# Ordenamiento por defecto
"Compr√© {} {} a ${} por un total de ${}".format(6, 'bananas', 500, 3000)
```

```python
# Ordenamiento especificado
"Por un total de ${3} compre {0} {1} a ${2} cada una.".format(6, 'bananas', 500, 3000)
```

```python
# Dar formato a los n√∫meros
print('Compr√© {:3d} {:8s} a ${:.3f} por un total de ${:.5f}'.
      format(6, 'bananas', 100.00, 600))
```

```python
# Ordenamiento especificado y formateo
print('Por un total de ${3:.5f} compr√© {0:3d} {1:8s} a ${2:.3f} '.
      format(6, 'bananas', 100.00, 600))
```

```python
# Porcentaje
'{:.2f}%'.format(10.12345)
```

```python
# Con palabras clave
print('Compr√© {cantidad} {item} a ${precio_unitario} por un total de ${total}'.
      format(cantidad=6, item='bananas', precio_unitario=100.00, total=600))
```

```python
# Declarar las variables por aparte
cantidad = 6
item = 'bananas'
precio_unitario = 100
total = cantidad * precio_unitario
print('Compr√© {a} {b} a ${c} por un total de ${d}'.format(a=cantidad, b=item, c=precio_unitario, d=total))
```

#### Ajuste del espacio

```python
# Agregar espacios u otro car√°cter a la izquierda hasta alcanzar 9 d√≠gitos
'{:>9}'.format(3.14)
```

```python
# Reservar espacio para 20 caracteres y alinear sobre el margen derecho
'{:>20} '.format('Python')
```

```python
# Reservar espacio para 20 caracteres y alinear sobre el margen izquierdo
'{:20} '.format('Python')
```

```python
# Reservar espacio para 20 caracteres y centrar
'{:^20} '.format('Python')
```
