#### Cadena f o f-string

```python
cant = 500
unit = 2.578
tot = cant * unit
f"Cant: {cant}, Precio unit.:{unit:7.3f}, Total:{tot:8.2f}"
```

```python
nombre = "Python"
v1 = 3
v2 = 9
print(f'En este curso se usa la versión {v1}.{v2} de {nombre}')
```

```python
# Operar con funciones
texto = "mayusculas"
print( f"Cambiar a mayusculas:  {texto.upper()}" )
```

```python
# Imprimir comillas usando \
nombre1 = 'Gustavo'
nombre2 = 'Augusto'
print(f'Buena cursada les deseamos \'{nombre1}\' y \'{nombre2}\'')
```

```python
# Con diccionarios usar comillas diferentes
docente = {'nombre1' : 'Gustavo', 'nombre2' : 'Augusto'}
print(f"Buena cursada les deseamos '{docente['nombre1']}' y '{docente['nombre2']}'")
```
