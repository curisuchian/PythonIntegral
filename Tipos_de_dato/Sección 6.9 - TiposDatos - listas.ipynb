Una lista (‘list’) es una colección ordenada de datos de idéntico o diferente tipo.

##### Crear listas

```python
# Empleando corchetes []
lista0 = [2018, 2019, 2020]                                
print(lista0)
type(lista0)
```

```python
lista1 = ['Almagro', 345, 'Pereda', 1900, 'Soprano', 888]  
print(lista1)
```

```python
# Empleando el constructor list con doble paréntesis. 
lista2 = list(('Almagro', 345, 'Pereda', 1900, 'Soprano', 888))
print(lista2)
```

##### Operadores

```python
a=[1,2,3,4]
b=[1,2,3,4]
c=[1,1,2,3]
print(a == b)   
print(a == c)   
```

##### Acceso a elementos

```python
lista = [2018, 2019, 2020]
print(lista[1])      
```

```python
# Remplazar valores
lista = [2018, 2019, 2020]

lista[1] = 2099  
print(lista)
```

```python
# Extracción usando tajada
lista = [2016, 2017, 2018, 2019, 2020]
lista[1:4]
```

#### Mutabilidad

```python
# Las listas son mutables
lista = [2016, 2018, 2019, 2020, 2020] 
lista.insert(1, 2017)        
print(lista)
```

```python
# Remplazo usando tajada
lista = [2016, 2017, 2018, 2019, 2020]
lista[0:3] = [2021, 2022, 2023]
print(lista)
```

##### Usos
Almacenar colecciones ordenadas de datos de modo tal que puedan ser modificadas durante la ejecución de un programa.
