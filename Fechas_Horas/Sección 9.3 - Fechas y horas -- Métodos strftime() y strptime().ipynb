#### Método strftime(format)
- Los objetos date, time y datetime tienen el método strftime(format).
- strftime devuelve una cadena representando la fecha según el formato en format.
- La tabla da un listado parcial de códigos de format. 

```python
from  datetime import date

hoy = date.today()    # Solo devuelve la fecha calendario
print(hoy)
print(hoy.strftime("%d/%B/%y"))  # Imprime segun formato
```

```python
from datetime import date

x = date(2018, 6, 30)
print(x.strftime("%d %B %Y"))  
```

```python
# Agrego %A
print(x.strftime("%A %d %B %Y")) 
```

```python
# Otros códigos
print(x.strftime("%A / %d / %b / %y"))  
```

##### Método strptime(cadena, formato)
Los objetos time y datetime tienen el método strptime().<br>
Crea un objeto datetime a partir de una cadena que representa fecha y hora de acuerdo a un formato.

```python
from datetime import datetime

año_nuevo =  "31 December, 2021"
año_nuevo2 = datetime.strptime(año_nuevo, "%d %B, %Y")
print("En documentos es mejor escribir la fecha en formato convencional: {}".format(año_nuevo2))
```

```python
from datetime import datetime

d = "12/31/2021 00:00:00"
fecha = datetime.strptime(d, "%m/%d/%Y %H:%M:%S")
print(fecha)
```

```python
# Ambigüedad
f = "4/6/2021"
f1 = datetime.strptime(f, "%m/%d/%Y")
f2 = datetime.strptime(f, "%d/%m/%Y")
print(f1)
print(f2)
```
