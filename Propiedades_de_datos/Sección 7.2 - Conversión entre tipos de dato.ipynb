El tipo de un dato puede ser modificado por Python o por el usuario.<br>
En el modo implícito lo hace Python <br>
En el modo explícito lo hace el usuario.

#### Modo implícito

```python
x = 5
y = 5.5
z = x + y
print(type(z))
```

```python
print(type(5 / 4))
```

#### Modo explícito

```python
a = '3.1415'
print(float(a))
```

```python
# Pérdida de información
x = 4.5
x = int(x)   
print(type(x))   # devuelve int; es decir, x fue convertido de float a integer. 
print(x)         # notar el truncamiento  
```

```python
int(False)
```

Se tiene un número entero x = 222 y una cadena representando un número entero y = ‘444’ (por ejemplo, devuelto por input()), y se desea sumar ambos.

```python
x = 222
y = '444'
x + y   # Arroja error aduciendo que el operador + no es válido con tipos mixtos
```

```python
# La solución es convertir la cadena en entero mediante coerción
x = 222
y = int('444')
print(x + y)
```

#### Tabla de códigos ASCII

```python
# Convertir a carácter la representación en ASCII de la letra b
print(type(98))
a = chr(98)
print(a)
print(type(a))
```

```python
# Hallar la representación en ASCII de la letra b
print(type('b'))
s = ord('b')
print(s)
print(type(s))
```
