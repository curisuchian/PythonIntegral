```python
# Transferir del ámbito local al global
def suma(x, y):
    f = x + y
    print(f)
    return f               # con return envia a al ambito global
    
f = suma(4, 6)             # fuera de la funcion
print(f)                   # ahora a es accesible para print()
```

El alcance se divide en cuatro niveles. <br>
__Alcance incorporado__ - es accesible desde cualquier sitio de un script <br>
__Alcance global__ – es accesible desde cuaquier sitio de un script <br>
__Alcance englobante__ - es accesible desde el interior de una función <br>
__Alcance local__ – es accesible desde la misma función <br>
