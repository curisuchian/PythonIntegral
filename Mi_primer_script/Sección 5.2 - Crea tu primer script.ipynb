```python
# Titulo: Conversor de temperatura
# Proposito: Convertir de grados Fahrenheit a grados Celsius
# Autor: GGB
# Fecha:  hoy
```

```python
def fahr_celsius(gradFahr):
    """ 
    Conversor de temperatura 
    Ingrese grados Fahrenheit
    Devuelve grados Celsius
    Emplea la expresion (gradFahr - 32) * 5/9
    """
    return (gradFahr - 32) * 5/9    # Expresion de calculo
    
a = fahr_celsius(100)            # Invocacion de la funcion
print(a)                # Impresion del resultado
```
