### Operadores

#### Operadores aritméticos

```python
print(1 + 5)
print(1 - 5)
print(1 * 5)
print(1 / 5)
```

#### Operadores de asignación

```python
a = 5
a
```

```python
i = 2
i += 5
i
```

```python
i = 2
i -= 1
i
```

#### Operadores relacionales o de comparación

```python
print('B' == 'B')
```

```python
print(2 < 3)
```

#### Operador de pertenencia

```python
iter = ['a', 'b', 'c']
print('a' in iter)
```

```python
iter = ['a', 'b', 'c']
print('a' not in iter)
```

#### Operadores lógicos

##### Operador and
Tabla de verdad

```python
print((1 < 2) and (2 < 3))
```

##### Operador or
Tabla de verdad

```python
print((1 > 2) or (2 > 3))
```

##### Operador not
Tabla de verdad

```python
print(not (2 > 3))
```
